```python
from tika import parser
import shutil
import os, sys, subprocess
import re
from glob import glob
```

```python
# 통화정책방향 혹은 통화신용정책운영방향을 안건으로 상정한 문서 분류

# pdf file 오류 확인
# !pip install tika
my_local_pdf_dir_path = 'MPB'
using_file=[]
using_file_path=[]
file_paths = glob(f'./{my_local_pdf_dir_path}/*.pdf')
for path in file_paths:
    raw = open(path, 'rb').read()
    try: data = parser.from_buffer(raw)
    except: data = parser.from_buffer(raw)
    content = data['content'].strip()
    content2 = re.sub(' |\n|\t', '', content)
    if re.findall("의안.+?통화정책방향", content2) or re.findall("의안.+?통화신용정책운영방향", content2):
        if int(data['metadata']['xmpTPg:NPages'])>6:
            using_file.append(content)
            using_file_path.append(path)
print('total length:',len(file_paths))
print(len(using_file), len(using_file_path))
```

```python
#비교할 문서제목을 정리
usings=[]
for i in using_file_path:
          i= i[6:]
          usings.append(i)

#문서비교
source_folder="./MPB/"
pdf_files = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]

#필터링을 통해 리스트 만들기 
filtered_list = list(filter(lambda x: x in usings, pdf_files))
filtered_list
```

```python
#pdf파일 txt로 변환
def pdf2txt(source_folder="./MPB/", output_folder="./MPB.text/"):
    
    # 지정 폴더 내 파일 목록 조회 (파일만)
    #print(pdf_files)

    for pdf in filtered_list:
        try:
            pdf_filepath = source_folder + pdf
            pdf_tmp_filepath = output_folder + 'tmp.pdf'

            # pdf 파일을 text로 변환
            shutil.copyfile(pdf_filepath, pdf_tmp_filepath)
            parsedPDF = parser.from_file(pdf_tmp_filepath)["content"]
            #print(parsedPDF)

            #enter 삭제 
    #             parsedPDF = re.sub('\n', '', parser.from_file(pdf_tmp_filepath)["content"])

            output_filepath = (output_folder + pdf).replace('.pdf', '.txt')
            #print(output_filepath)
            with open(output_filepath, 'w',-1,'utf-8') as f:
                print(output_filepath)
                f.write(parsedPDF)
                f.close()
        except:
            print('Error Occurred at {}:'.format(pdf))
            print("Error:", sys.exc_info()[0])
            pass
```
