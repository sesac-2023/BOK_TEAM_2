```python
#콜금리 데이터 크롤링

import time
import datetime
import requests, os
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm

last_url = f"https://www.korcham.net/nCham/Service/EconBrief/appl/ProspectBoardList.asp?board_type=1&pageno=1&daybt=OldNow&m_OldDate=20200201&m_NowDate=20230901"
call_rates=[]
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.1938.69'}
response = requests.get("https://www.korcham.net/nCham/Service/EconBrief/appl/ProspectBoardList.asp?board_type=1&pageno=1&daybt=OldNow&m_OldDate=20130101&m_NowDate=20230901", headers=headers)
soup = BeautifulSoup(response.text)
last_page = str(soup.select('.next')[-1]).split('\'')[1]


for page_num in range(0,int(last_page)):
    response = requests.get(f"https://www.korcham.net/nCham/Service/EconBrief/appl/ProspectBoardList.asp?board_type=1&pageno={page_num+1}&daybt=OldNow&m_OldDate=20130101&m_NowDate=20230901", headers=headers)
    soup = BeautifulSoup(response.text)
    call_rate = soup.select("tbody > tr > td")
    for i in call_rate:
        call_rates.append(i.get_text())
    
print(call_rates)
```

```python
#필요한 데이터만 정리: 날짜 및 일별 콜금리
def divide_list(l, n): # 리스트 l의 길이가 n이면 계속 반복
    for i in range(0, len(call_rates), n): 
        yield l[i:i + n] 

n = 6 # 한 리스트에 몇개씩 담을지 결정
day_callrates = list(divide_list(call_rates, n))
print(day_callrates)


day_Callrate= list(zip(day_callrates[:][0],day_callrates[:][1])) #날짜, 일별콜금리만 list 안에 넣기
```

```python
#필요한 데이터만 정리: 리스트에 넣기
day_Callrates=[]
for i in range(len(day_callrates)):
        print(day_callrates[i][0:2])
        day_Callrates.append(day_callrates[i][0:2])
```

```python
#데이터프레임화, csv파일로 저장
import pandas as pd
df = pd.DataFrame(day_Callrates, columns=['date','call_rate'])
df= df.sort_values('date')
df['date'] = df['date'].str.replace('.','-')
df =df.reset_index(drop=True)
df.to_csv("callrates_2013to2023.csv", index=None)
```

```python
#날짜 형식으로 바꾸기
from datetime import datetime 
import time
df['date']= pd.to_datetime(df['date'])
```

```python
#레이블링: polarity score 계산을 위해 기준지표인 '한달 전 금리' 데이터 생성
df = df.merge(df.assign(date = df.date + pd.Timedelta(days=30)),
                            on= 'date',
                            how='left', suffixes = ['','_30daysago'])
df.dropna(inplace=True)
df['call_rate'] = df['call_rate'].astype(float)
df['call_rate_30daysago'] = df['call_rate_30daysago'].astype(float)
```

```python
#레이블링: '한달 전 금리' 를 기준으로 일별 콜금리 변화 방향을 측정. 이를 토대로 레이블링 
import numpy as np
df['change'] = np.select(
    condlist=[
        (df.call_rate > df.call_rate_30daysago),
        (df.call_rate == df.call_rate_30daysago),
        ],
        choicelist=[
            1,
            0,
        ],
        default=-1
)
```
